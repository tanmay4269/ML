```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

### Simple example

```python
x_axis = np.arange(4)
y_axis = np.random.randint(1,10,size=(4))

print(x_axis)
print(y_axis)
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)

ax.set_xlabel("Time")
ax.set_ylabel("Position")
```

```python
b = np.matrix([[1, 2], [3, 4]])
b_asarray = np.asarray(b)
print(b)
print(b_asarray)
```

### Example 2

```python
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots()
ax.plot(x, x**2, label="quad")
ax.plot(x, x**3, label="cubic")
ax.legend()
```

### Alternate method to do the same as above

```python
x = np.linspace(0, 5, 100)

plt.figure()
plt.plot(x, np.exp(x), label="expo")
plt.legend()
```
