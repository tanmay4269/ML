```python
import matplotlib.pyplot as plt
import numpy as np
```

### Simple example

```python
n = 10

x = np.arange(n)
y = np.random.randint(0, n, size=(n))

# resize graph
plt.figure(figsize=(6,4), dpi=150)

plt.plot(x, y, 
        color="r", linestyle="-.", linewidth=2, # line styling
        marker="o", markeredgecolor="orange", markersize=10,  # marker styling
        label="random")

# short hand notation: [color][marker][line]
plt.plot(x, np.abs(np.sin(x))*10, "bx--", label="full-rectified")

plt.title("Nice plot :)", fontdict={"fontname": "roboto"})
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.xticks(x)
plt.yticks(x)

plt.legend()
plt.savefig("Nice_plot.png", dpi=300)  # you can save fig in a better resolution than whats rendered
plt.show()
```

### Bar charts

```python
labels = ['A', 'B', 'C']
values = [1,4,3]

bars = plt.bar(labels, values)

bars[0].set_hatch('/')
bars[1].set_hatch('.')
bars[2].set_hatch('O')

plt.show()
```
