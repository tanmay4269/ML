# Fourier series

```python
import numpy as np 
import matplotlib.pyplot as plt 

plt.figure(figsize=(7,2), dpi=150) 

PI = 3.14

n = 3  # frequency for raw wave (for example purpose its a pure wave)

t = np.linspace(0, 4, 10**4)
raw_wave = (1 + np.cos(2*PI*n*t)) / 2


plt.plot(t, raw_wave)
plt.show()
```
